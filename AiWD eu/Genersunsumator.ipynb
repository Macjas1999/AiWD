```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Genersuno.csv')
```

```python
drop0 = data['sum'] > 0
```

```python
gnum = data.loc[drop0 ,['sum', 'year']]
```

```python
gnum
```

```python
plt.style.use('ggplot')

x = gnum['year']

y = gnum['sum']

x_ind = np.arange(len(x))

plt.barh(x_ind, y,0.5,100, color='#4b4e4e', linestyle='--')


plt.yticks(x_ind, x)

plt.xlabel('Liczba samobójstw')
plt.ylabel('Lata (z pominięciem lat bez danych)')
plt.title('Samobójstwa w Niemczech')

#plt.legend()  label='All Devs'

plt.tight_layout()

plt.savefig('jp.png')

plt.show()
```
