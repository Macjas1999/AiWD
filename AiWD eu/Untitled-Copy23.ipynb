```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('PolandssS1.csv')
```

```python
drop0 = data['sum'] > 0
```

```python
data = data.loc[drop0,['sum', 'year']]
```

```python
data
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10 ,6))

x = data['year']


y = data['sum']


plt.plot(x, y, color= 'r', linestyle='--', marker='o', label='Liczba samobójstw')



plt.xlabel('Lata')
plt.ylabel('Liczba samobójstw')
plt.title('Samobójstwa: Polska')

plt.legend()

plt.tight_layout()
plt.savefig('Pol.png')

plt.show()
```
