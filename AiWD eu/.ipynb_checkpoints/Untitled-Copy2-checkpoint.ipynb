```python
plt.style.use('ggplot')
plt.figure(figsize=(10 ,6))

x = data['year']


yF = data['sumF']


plt.plot(x, yF, color= 'r', linestyle='--',label='Kobiety')

yM = data['sumM']

plt.plot(x, yM, color= 'b', linestyle='--',label='Mężczyźni')



plt.xlabel('Lata')
plt.ylabel('Liczba samobójstw')
plt.title('Samobójstwa: kobiety vs mężczyźni')

plt.legend()

plt.tight_layout()
plt.savefig('MvsK.png')

plt.show()
```
