```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
data = pd.read_csv('masterCSVfmfull.csv')
```

```python
agg_pop = data.groupby('year')['population'].agg(['sum', 'mean', 'count'])
agg_pop = agg_pop.reset_index()
```

```python
ftl1= (agg_pop['year'] != 2016)
agg_pop = agg_pop.loc[ftl1, ['year', 'sum', 'mean','count']]
```
