```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('AustriassS.csv')
```

```python
ftl = (data['sum'] > 0)
```

```python
df = data.loc[ftl]
df
```

```python
plt.style.use('fivethirtyeight')


x = df['year']

y = df['sum']

plt.plot(x,y, label='gr')



plt.xlabel('Rok')
plt.ylabel('Liczba samobójstw')
plt.title('Liczba samobójstw na przestrzeni lat')

plt.legend()

plt.tight_layout()

#plt.savefig('Albgraph')

plt.show()
```
