```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_sb = pd.read_csv('agg_mxminct.csv')
data_sbpkp = pd.read_csv('agg_mxminct_100kpoprst.csv')
```

```python
data_sb = data_sb.drop(['Unnamed: 0'], axis = 1)

data_sbpkp = data_sbpkp.drop(['Unnamed: 0'], axis = 1)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12 ,8))

x = data_sb['year']


y1 = data_sb['mean']

plt.subplot(211)
plt.plot(x, y1, color= '#65374D', linestyle='--', marker='o', label='Srednia samobójstw')
plt.xlabel('Lata')
plt.ylabel('Wartość średniej')
plt.title('Średnia ilość samobójstw na świecie')
plt.legend()
plt.tight_layout()

y2 = data_sbpkp['mean']

plt.subplot(212)
plt.plot(x, y2, color= '#000034', linestyle='--', marker='o', label='średnia samobójstw na 100 tyś. mieszkańców')



plt.xlabel('Lata')
plt.ylabel('Wartość średniej')
plt.title('Średnia wartość stosunku liczby samobójstw na sto tysięcy mieszkcńców na świecie')

plt.legend()

plt.tight_layout()
plt.savefig('Średnia ilość samobójstw na świecie.png')

plt.show()
```
