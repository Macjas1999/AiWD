```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
data = pd.read_csv('masterCSVfmfull.csv')
```

```python
agg_mxminct = data.groupby('year')['suicides_no'].agg(['max', 'mean', 'count'])
agg_mxminct = agg_mxminct.reset_index()
```

```python
ftl1= (agg_mxminct['year'] != 2016)
agg_mxminct = agg_mxminct.loc[ftl1, ['year', 'max', 'mean','count']]
```

```python
agg_mxminct
```

```python
agg_mxminct.to_csv('agg_mxminct.csv')
```

```python
agg_mxminct_100kpop = data.groupby('year')['suicides/100k pop'].agg(['max', 'mean', 'count'])
```

```python
agg_mxminct_100kpoprst = agg_mxminct_100kpop.reset_index()
```

```python
ftl= (agg_mxminct_100kpoprst['year'] != 2016)
```

```python
agg_mxminct_100kpoprst = agg_mxminct_100kpoprst.loc[ftl, ['year', 'max', 'mean','count']]
```

```python
agg_mxminct_100kpoprst.to_csv('agg_mxminct_100kpoprst.csv')
```
