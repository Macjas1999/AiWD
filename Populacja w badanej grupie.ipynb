```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
agg_pop = pd.read_csv('agg_pop.csv')
```

```python
agg_pop = agg_pop.drop(['Unnamed: 0'], axis = 1)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12 ,8))

x = agg_pop['year']


y = agg_pop['sum']

plt.plot(x, y, color= '#009cb2', linestyle='--', marker='o',)

plt.xlabel('Lata')
plt.ylabel('Liczebność populacji wyrażona w postaci y*10^9')
plt.title('Liczebność populacji badanych państw na przestrzeni obserwowanego czasu')


plt.tight_layout()
plt.savefig('Liczebność populacji badanych państw na przestrzeni obserwowanego czasu.png')

plt.show()
```
