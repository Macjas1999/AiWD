```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Genersuno.csv')
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8 ,8))

x = data['agerange']

yF = data['female']

x_ind = np.arange(len(x))

nms = ['5-14', '15-24', '25-34', '35-54', '55-74', '75+']

width = 0.15

plt.bar(x_ind + width , yF,width=0.25, color='r', label='Kobiety')

yM = data['male']

plt.bar(x_ind - width, yM,width=0.25, color='b',label='Mężczyźni')

plt.xticks(x_ind, nms)


plt.xlabel('Przedział wiekowyw (w latach)')
plt.ylabel('Liczba samobójstw')
plt.title('Liczba samobójstw względem przedziału wiekowego na Świecie')

plt.legend()

plt.tight_layout()
plt.savefig('Wiek_mf.png')

plt.show()
```
